### 2.1.2 Variable Types

#### Local Variable

Local variables are declared inside a function or block and are accessible only within that specific scope. They are temporary and are destroyed once the function exits.


**Characteristics**:
* Created when the function starts execution.
* Accessible only within the function.
* Cannot be accessed outside the function.

````python
def my_function():
    x = 10  # Local variable
    print("Inside function:", x)

my_function()
# print(x)  # Error: x is not defined outside the function
````

#### Global Variables

Global variables are declared outside any function, class, or block. They are accessible throughout the entire program, including inside functions.

**Characteristics**:
* Defined at the top level of a script.
* Accessible inside and outside functions unless shadowed by a local variable.
* Can be modified inside a function using the global keyword.

````python
x = 20  # Global variable

def my_function():
    global x
    x = 30  # Modifying global variable
    print("Inside function:", x)

my_function()
print("Outside function:", x)  # Outputs 30
````

Without the global keyword, reassigning x inside the function creates a new local variable instead of modifying the global one.

````python
x = 20

def my_function():
    x = 30  # Local variable; does not affect the global x
    print("Inside function:", x)

my_function()
print("Outside function:", x)  # Outputs 20
````

#### Constants

Constants are variables whose values should not change during the program's execution. Python does not have built-in constant support, but constants are typically declared using uppercase variable names as a convention.

**Characteristics:**
* Declared using uppercase letters.
* Should not be modified by convention, but technically can be (Python does not enforce immutability).

````python
PI = 3.14159  # Constant
GRAVITY = 9.8

def calculate_circumference(radius):
    return 2 * PI * radius

print(calculate_circumference(5))  # 31.4159
````

Attempting to modify a constant:

````python
PI = 3.14159
PI = 3.14  # No error, but violates the convention
````

For true immutability, you can use libraries like enum or dataclasses.

#### Non-local Variable

Nonlocal variables are used inside nested functions to refer to variables in the nearest enclosing scope (excluding the global scope).

Characteristics:
* Declared using the nonlocal keyword.
* Useful for modifying variables in the enclosing scope within a nested function.

````python
def outer_function():
    x = 10  # Enclosing variable

    def inner_function():
        nonlocal x
        x += 5  # Modify the enclosing variable
        print("Inner function:", x)

    inner_function()
    print("Outer function:", x)

outer_function()
# Output:
# Inner function: 15
# Outer function: 15
````

***Note : Without nonlocal, the variable x inside inner_function would be treated as local to it.***

#### Scope

The scope of a variable determines the region of the code where it is accessible. Python has four types of variable scopes, defined by the LEGB rule:

* Local: Variables inside a function or block.
* Enclosing: Variables in the nearest enclosing function.
* Global: Variables at the module level.
* Built-in: Predefined names in Python

````python
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print("Inner:", x)

    inner()
    print("Outer:", x)

outer()
print("Global:", x)
````

#### 5.8.2 Operator Overloading

Allows operators to have user-defined meanings for custom objects.


````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)

p1 = Point(1, 2)
p2 = Point(3, 4)
result = p1 + p2
print(result.x, result.y)  # Outputs: 4, 6
````

## 6.Control Flow

Control flow in Python refers to the order in which individual statements, instructions, or function calls are executed or evaluated. Python provides several control flow mechanisms to direct the programs flow based on conditions or iterations.

### 6.1 Conditional Statements

Conditional statements allow code execution based on certain conditions.

#### If Statements

Executes a block of code if the condition is True.

````python
x = 10
if x > 5:
    print("x is greater than 5")  # Output: x is greater than 5
````

#### if-else Statement

Provides an alternative block of code if the condition is False.

````python
x = 3
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")  # Output: x is not greater than 5
````

#### if-elif-else Statement

Checks multiple conditions sequentially.

````python
x = 7
if x > 10:
    print("x is greater than 10")
elif x > 5:
    print("x is greater than 5 but less than or equal to 10")  # Output
else:
    print("x is less than or equal to 5")
````

### 6.2 Loops

#### for Loop

Iterates over a sequence (list, tuple, dictionary, set, or string).


````python
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
    print(fruit)
# Output:
# apple
# banana
# cherry
````

Using range() in for loops:

````python
for i in range(1, 6):
    print(i)
# Output: 1, 2, 3, 4, 5
````

#### while Loop

Repeats as long as the condition is True.

````python
count = 0
while count < 5:
    print(count)
    count += 1
# Output: 0, 1, 2, 3, 4
````

### 6.3 Control Statements in Loops

These control the behavior of loops.

#### Break

Exits the loop prematurely.

````python
for i in range(10):
    if i == 5:
        break
    print(i)
# Output: 0, 1, 2, 3, 4
````

#### continue

Skips the current iteration and continues with the next.

````python
for i in range(5):
    if i == 2:
        continue
    print(i)
# Output: 0, 1, 3, 4
````

#### pass

Does nothing; used as a placeholder.

````python
for i in range(3):
    pass  # Placeholder for future code
````

#### else with Loops

The else block executes if the loop completes normally (without break).

````python
for i in range(5):
    print(i)
else:
    print("Loop completed successfully")
# Output: 0, 1, 2, 3, 4
#         Loop completed successfully
````

### 6.4 Match-Case Statement (Python 3.10+)

Introduced as a replacement for switch-case in other languages.

````python
day = "Monday"
match day:
    case "Monday":
        print("Start of the week")
    case "Friday":
        print("Almost weekend")
    case _:
        print("It's a regular day")
# Output: Start of the week
````

### 6.5 Exception Handling Control Flow

Exception handling ensures graceful handling of errors.

#### Try-Except

Handles specific or generic exceptions.

````python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
# Output: Cannot divide by zero
````

#### Try-Except-Finally

The finally block always execution

````python
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Error occurred")
finally:
    print("Execution complete")
# Output: Execution complete
````

#### Try-Except-Else

Executes the else block if no exceptions occur.

````python
try:
    result = 10 / 2
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("No errors occurred:", result)  # Output: No errors occurred: 5.0
````

### 6.6 Comprehensions with Control Flow

List comprehensions allow compact loop and conditional logic.

````python
numbers = [1, 2, 3, 4, 5]
squared = [x**2 for x in numbers if x % 2 == 0]
print(squared)
# Output: [4, 16]
````

### 6.7 Nested Control Flow

Control flow can be nested within loops or conditional statements.

````python
for i in range(1, 4):
    for j in range(1, 4):
        print(f"i={i}, j={j}")
# Output:
# i=1, j=1
# i=1, j=2
# i=1, j=3
# i=2, j=1
# ...
````

### 6.8 Infinite Loops

These occur if the exit condition is never met.

````python
while True:
    user_input = input("Enter 'exit' to stop: ")
    if user_input == "exit":
        break
````

### 6.9 Control Flow for Functions

#### Return Statement

Terminates the function and optionally returns a value.

````python
def add(a, b):
    return a + b

print(add(3, 4))  # Output: 7
````

### 6.10 Control Flow for Generators

Generators yield values one at a time.

````python
def generate_numbers():
    for i in range(1, 4):
        yield i

for num in generate_numbers():
    print(num)
# Output: 1, 2, 3
````

## 7.Functions

A function is a block of organized, reusable code that performs a single, related action. Functions make the code more manageable, testable, and reusable.

````python
# Synatx
def function_name(parameters):
    """
    Optional Docstring: Explains what the function does.
    """
    # Function body
    return expression  # Optional return statement
````

### 7.2 Function Arguments and Parameters

#### Positional Arguments

Arguments matched to parameters in the order they are passed.

````python
def subtract(a, b):
    return a - b

print(subtract(10, 5))  # Output: 5
````

#### Default Arguments

Parameters with default values. If no value is provided, the default is used.

````python
def greet(name="Guest"):
    return f"Hello, {name}!"

print(greet())         # Output: Hello, Guest!
print(greet("Alice"))  # Output: Hello, Alice!
````

#### Keyword Arguments

Arguments passed using parameter names, regardless of their order.

````python
def divide(a, b):
    return a / b

print(divide(b=10, a=20))  # Output: 2.0
````

 #### Variable-Length Arguments

Used when the number of arguments is unknown.

*args (Non-Keyworded Arguments)

Accepts a variable number of positional arguments as a tuple.

````python
def sum_all(*args):
    return sum(args)

print(sum_all(1, 2, 3, 4))  # Output: 10
````

**kwargs (Keyworded Arguments)

Accepts a variable number of keyword arguments as a dictionary.

````python
def display_info(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

display_info(name="Alice", age=25, city="New York")
# Output:
# name: Alice
# age: 25
# city: New York
````

### 7.3 Function Decorators

A decorator is a function that modifies another function or method. It is often used for logging, enforcing access control, or measuring performance.

````python
def decorator(func):
    def wrapper():
        print("Before the function call")
        func()
        print("After the function call")
    return wrapper

@decorator
def say_hello():
    print("Hello!")

say_hello()
# Output:
# Before the function call
# Hello!
# After the function call
````

### 7.4 Recursive Functions

A function that calls itself to solve smaller instances of a problem.

````python
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))  # Output: 120
````

### 7.5 Function Annotations

Annotations provide metadata for parameters and return types.

````python
def add(a: int, b: int) -> int:
    return a + b

print(add(3, 4))  # Output: 7
````

### 7.6 Higher-Order Functions

Functions that take other functions as arguments or return them.

````python
def apply_function(func, value):
    return func(value)

print(apply_function(lambda x: x**2, 3))  # Output: 9
````

### 7.7 Generators

Functions that yield values using the yield keyword instead of return.

````python
def generate_numbers():
    for i in range(1, 4):
        yield i

for num in generate_numbers():
    print(num)
# Output: 1, 2, 3
````

## 8.Modules and Packages

Modules and packages are essential concepts in Python that help organize and structure code effectively. They promote code reusability, maintainability, and readability, especially in large projects.

### 8.1 Modules

A module is a single Python file that contains code (functions, variables, classes) which can be reused in other programs. Modules allow logical organization of Python code.

#### Creating a Module

We can create a module by saving a .py file.

Example:
Create a file math_operations.py:

````python
# math_operations.py
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b
````

#### Using a Module

We can import a module in another Python file using the import statement.

````python
# main.py
import math_operations

print(math_operations.add(5, 3))        # Output: 8
print(math_operations.subtract(10, 4))  # Output: 6
````

#### Importing Specific Functions

We can import specific items from a module using the from ... import syntax.

````python
from math_operations import add

print(add(4, 5))  # Output: 9
````

#### Importing with Aliases

We can use an alias for a module using the as keyword.

````python
import math_operations as mo

print(mo.add(7, 2))  # Output: 9
````

#### Built-in Modules

Python includes many pre-installed modules. Examples: math, os, sys, random, datetime.

````python
import math
print(math.sqrt(16))  # Output: 4.0
````

#### Custom Modules

Custom modules are those created by the user. For example, the math_operations.py above is a custom module.

##### __name__ and __main__

Modules have a special variable __name__ that is set to "__main__" when the module is executed directly. If the module is imported, __name__ is set to the module's name.

````python
# module_example.py
def main():
    print("Executed as a script.")

if __name__ == "__main__":
    main()
````

###### Custom Objects with iter()

If you create a custom class, you can define the __iter__() method to make it iterable

````python
class Counter:
    def __init__(self, limit):
        self.limit = limit
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.limit:
            self.count += 1
            return self.count
        else:
            raise StopIteration

counter = Counter(3)
for number in counter:
    print(number)
````

###### Using iter() with Sentinel Value

iter() can also be used with a callable object and a sentinel value. In this mode, iter(callable, sentinel) repeatedly calls the callable until the sentinel value is returned.

````python
def counter():
    i = 0
    while True:
        i += 1
        yield i

iterator = iter(counter().__next__, 5)

for value in iterator:
    print(value)
````

**Description:**

* If there are more items in the iterator, next() returns the next item.
* If the iterator is exhausted, and no default is provided, it raises a StopIteration exception.

````python
iterator = iter([10, 20, 30])

print(next(iterator))  # Output: 10
print(next(iterator))  # Output: 20
print(next(iterator))  # Output: 30

# Using default value to avoid StopIteration
print(next(iterator, 'No more items'))  # Output: No more items
````

### 11.8 Memory and Object Handling


Pythons memory and object handling mechanism ensures efficient memory utilization while providing developers with tools to manage data structures seamlessly. It involves concepts such as memory allocation, garbage collection, reference counting, and object lifecycle management.

#### Memory Management

##### Memory Allocation

Python uses private heap space for memory management, where all objects and data structures are stored.

* Python Interpreter: Allocates memory for objects and handles garbage collection automatically.
* Dynamic Memory Allocation: Python objects are created dynamically as needed.
* Memory Pooling: Python reuses memory for frequently used objects to optimize performance.

##### Memory Layout

* Stack: Stores function calls and local variables.
* Heap: Stores dynamically allocated objects such as lists, dictionaries, and user-defined objects.
* Global Memory: Stores global variables and module-level objects.

#### Object Lifecycle

Every object in Python has a lifecycle that includes:

* Creation: Objects are created in memory when declared.
* Manipulation: Objects are used and modified during the program execution.
* Destruction: Objects are removed from memory when no longer needed.

#### Reference Counting and Garbage Collection

##### Reference Counting

Python uses reference counting to manage object memory. Each object keeps a count of references pointing to it.

* Reference Count:
Increases when a new reference is created.
Decreases when a reference is deleted or goes out of scope.

* Object Deletion: If the reference count drops to zero, the object becomes eligible for garbage collection.

````python
a = [1, 2, 3]  # Reference count of the list is 1
b = a           # Reference count increases to 2
del a           # Reference count decreases to 1
del b           # Reference count decreases to 0, object is deleted
````

##### Garbage Collection

Pythons garbage collector handles objects that are no longer reachable. It removes circular references and reclaims memory.

* Circular References: Occurs when two or more objects refer to each other, forming a loop.
* gc Module: Provides control over garbage collection.
    * gc.collect(): Forces garbage collection.
    * gc.get_threshold(): Retrieves the garbage collection thresholds.

````python
import gc
class Node:
    def __init__(self, value):
        self.value = value
        self.ref = None

a = Node(1)
b = Node(2)
a.ref = b
b.ref = a  # Circular reference
del a, b   # Objects still exist in memory due to circular reference
gc.collect()  # Manually clears the circular reference
````

##### Small Integer and String Caching

* Integers: Python caches integers between -5 and 256 for reuse.
* Strings: Python caches short strings and commonly used literals.

````python
a = 256
b = 256
print(a is b)  # Output: True (points to the same object)

a = 257
b = 257
print(a is b)  # Output: False (different objects)
````

## 4.Generators & Decorators

### Generators

Generators are special types of iterators that yield values one at a time using the yield keyword instead of returning them all at once. This makes generators memory-efficient for handling large datasets or infinite sequences.

**Defining Custom Generators with yield:**

A generator function contains one or more yield statements. When called, it returns a generator object that can be iterated over.

````python
def count_up_to(n):
    count = 1
    while count <= n:
        yield count
        count += 1

gen = count_up_to(5)
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2
````

Key Characteristics of Generators:

- State Preservation: Generators preserve their state between calls.
- Lazy Evaluation: Values are produced only when needed, saving memory.
- Iteration Protocol: Generators are iterable and can be used in loops.

````python
# Example with Infinite Generator:

def infinite_numbers():
    num = 0
    while True:
        yield num
        num += 1

for number in infinite_numbers():
    if number > 10:
        break
    print(number)
````

### Decorators

Decorators are functions or classes that modify the behavior of other functions or methods. They are commonly used for logging, validation, or enforcing rules.

**Defining a Custom Decorator**

A decorator takes a function as input and returns a modified function.

````python
def my_decorator(func):
    def wrapper():
        print("Something is happening before the function is called.")
        func()
        print("Something is happening after the function is called.")
    return wrapper

@my_decorator
def say_hello():
    print("Hello!")

say_hello()
# Output:
# Something is happening before the function is called.
# Hello!
# Something is happening after the function is called.
````

 Defines a method that doesn't operate on an instance or class but logically belongs in the class.

### Linked Lists

**Explanation**

A linked list is a linear data structure where elements (nodes) are connected using pointers. Each node contains:

- Data.
- A pointer to the next node.

**Types**

- Singly Linked List: Each node points to the next.
- Doubly Linked List: Nodes have pointers to both previous and next nodes.
- Circular Linked List: The last node points back to the first node.

**Implementation**

Python does not have a native linked list; it can be implemented using classes.

**Use Cases**

- Dynamic memory allocation.
- Implementation of stacks and queues.

````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def print_list(self):
        current = self.head
        while current:
            print(current.data, end=" -> ")
            current = current.next
        print("None")

# Usage
ll = LinkedList()
ll.append(1)
ll.append(2)
ll.append(3)
ll.print_list()  # Output: 1 -> 2 -> 3 -> None
````

**Representation**
- Adjacency Matrix: 2D array representing connections.
- Adjacency List: Dictionary mapping nodes to neighbors.

````python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        if u not in self.graph:
            self.graph[u] = []
        self.graph[u].append(v)

    def print_graph(self):
        for node, neighbors in self.graph.items():
            print(f"{node} -> {', '.join(map(str, neighbors))}")

# Usage
g = Graph()
g.add_edge(1, 2)
g.add_edge(1, 3)
g.print_graph()  # Output: 1 -> 2, 3
````

### Tries

**Explanation**

A trie is a tree-like structure used for efficient string operations like prefix searching.

````python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end_of_word

# Usage
trie = Trie()
trie.insert("apple")
print(trie.search("apple"))  # Output: True
print(trie.search("app"))    # Output: False
````

- Tasks are used to run coroutines concurrently.

## 3.Global Interpreter Lock (GIL)

The Global Interpreter Lock (GIL) is a mutex in CPython that allows only one thread to execute Python bytecode at a time. This restricts Python's multithreading capabilities for CPU-bound tasks but does not affect I/O-bound tasks.

**Implications:**

- CPU-bound tasks: Limited performance improvement with threading.
- I/O-bound tasks: Effective, as the GIL is released during I/O operations.

## 4.Multiprocessing for Parallel Execution

For CPU-bound tasks requiring true parallelism, the multiprocessing module is used. It spawns separate processes, each with its own Python interpreter and memory space, bypassing the GIL.

````python
from multiprocessing import Process

def worker_function(name):
    print(f"Worker {name} is running")

if __name__ == "__main__":
    processes = [Process(target=worker_function, args=(i,)) for i in range(4)]
    for process in processes:
        process.start()
    for process in processes:
        process.join()
````

**Key Concepts in Multiprocessing:**

- Process: Represents an independent process.
- Pool: Manages multiple worker processes efficiently.
- Shared Memory: Allows sharing data between processes.

## 5.Combining Asynchronous and Multiprocessing

For complex applications, combining asyncio for I/O-bound tasks with multiprocessing for CPU-bound tasks can be effective.

````python
import asyncio
from multiprocessing import Process

async def async_task():
    print("Starting async task")
    await asyncio.sleep(2)
    print("Async task completed")

def cpu_task():
    print("Starting CPU-bound task")
    for _ in range(5_000_000):
        pass  # Simulates CPU work
    print("CPU-bound task completed")

if __name__ == "__main__":
    p = Process(target=cpu_task)
    p.start()

    asyncio.run(async_task())
    p.join()
````

**How CPython Works**

- Source Code Compilation:
    - Python source code (.py) is first compiled into bytecode (.pyc).
    - Bytecode is a low-level, platform-independent representation of the code.

- Execution via Python Virtual Machine (PVM):
    - The PVM reads and executes the bytecode instructions.

- Components of CPython:
    - Parser: Converts source code into an Abstract Syntax Tree (AST).
    - Compiler: Transforms AST into bytecode.
    - Interpreter: Executes the bytecode instructions.

### b.Key Features of CPython

- Global Interpreter Lock (GIL):
     - Ensures that only one thread executes Python bytecode at a time.
     - Simplifies memory management but limits multi-threading performance.

- Reference Counting:

    - Manages memory allocation and garbage collection.
    - Objects are deallocated when their reference count drops to zero.

### c.Advantages and Limitations of CPython

**- Advantages:**

- Compatibility with C libraries.
- Comprehensive standard library.

-** Limitations:**

- Slower execution compared to some compiled languages.
- GIL hampers multi-threading performance.

#### a.Creating Classes Dynamically

The type() function can create classes dynamically.

**Syntax:**
>type(name, bases, dict)

````python
# Dynamically creating a class
MyClass = type('MyClass', (object,), {'greet': lambda self: "Hello!"})

# Using the dynamically created class
obj = MyClass()
print(obj.greet())  # Output: Hello!
````

#### b.Creating Functions Dynamically

Functions can also be created dynamically using the exec() function or types.FunctionType.

**Example with exec():**

````python
code = """
def dynamic_function(x):
    return x * 2
"""
exec(code)
print(dynamic_function(5))  # Output: 10
````

**Example with types.FunctionType:**

````python
import types

def custom_body(x):
    return x + 1

dynamic_func = types.FunctionType(custom_body.__code__, globals())
print(dynamic_func(5))  # Output: 6
````

### 2.Metaclasses and the type() Function

A metaclass controls the behavior of a class and defines how a class behaves. In Python, classes themselves are instances of a metaclass, with type being the default metaclass.

#### **a. What is a Metaclass?**

- A class of a class.
- Determines how a class is created and initialized.

#### b.Creating a Metaclass

````python
class MyMeta(type):
    def __new__(cls, name, bases, dct):
        print(f"Creating class {name}")
        dct['greet'] = lambda self: "Hello from metaclass!"
        return super().__new__(cls, name, bases, dct)

class MyClass(metaclass=MyMeta):
    pass

obj = MyClass()
print(obj.greet())  # Output: Hello from metaclass!
````

### Celery

Celery is a distributed task queue system that enables asynchronous task execution. It's designed to handle real-time processing and background job execution.

**Key Features of Celery**

- Broker Integration: Uses brokers like RabbitMQ, Redis, or Amazon SQS for message passing.

- Task Scheduling: Supports periodic tasks via schedulers like celery-beat.

#### Core Components of Celery

**a. Broker**

A broker facilitates communication between the Celery client and workers. Redis and RabbitMQ are popular choices.

**b. Workers**

Workers execute tasks asynchronously. Multiple workers can run on different machines to process tasks in parallel.

**c. Task**

A task is a Python function decorated with @celery.task. These tasks are executed asynchronously.

**d. Result Backend**

Used to store task results for later retrieval.

### Graph Algorithms

##### Dijkstra's Algorithm (Shortest Path)

Dijkstras algorithm finds the shortest path from a starting node to all other nodes in a graph with non-negative edge weights.

**Step-by-Step Explanation:**

1. Initialize the distance to the start node as 0 and all other nodes as infinity.
2. Mark all nodes as unvisited.
3. Visit the unvisited node with the smallest known distance.
4. For each neighboring node, calculate the tentative distance and update if the new path is shorter.
5. Repeat until all nodes have been visited.

**Pseudocode:**
```
function dijkstra(graph, start):
    create a distance dictionary with infinity for all nodes
    set distance[start] = 0
    create a priority queue with (distance[start], start)
    
    while priority queue is not empty:
        current_distance, current_node = pop from queue
        
        if current_distance > distance[current_node]:
            continue
            
        for each neighbor in graph[current_node]:
            new_distance = current_distance + weight of edge
            if new_distance < distance[neighbor]:
                update distance[neighbor] = new_distance
                push (new_distance, neighbor) into priority queue

    return distance
```

````python
import heapq

def dijkstra(graph, start):
    distances = {vertex: float('infinity') for vertex in graph}
    distances[start] = 0
    priority_queue = [(0, start)]

    while priority_queue:
        current_distance, current_vertex = heapq.heappop(priority_queue)

        if current_distance > distances[current_vertex]:
            continue

        for neighbor, weight in graph[current_vertex].items():
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))

    return distances
````

### Sorting Algorithms

Sorting is a basic building block that many other algorithms are built upon. Its related to several exciting ideas that youll see throughout your programming career. We can use sorting to solve a wide range of problems:

**Measuring Efficiency With Big O Notation**

The specific time an algorithm takes to run isnt enough information to get the full picture of its time complexity. To solve this problem, you can use Big O (pronounced big oh) notation. Big O is often used to compare different implementations and decide which one is the most efficient, skipping unnecessary details and focusing on whats most important in the runtime of an algorithm.

The time in seconds required to run different algorithms can be influenced by several unrelated factors, including processor speed or available memory. Big O, on the other hand, provides a platform to express runtime complexity in hardware-agnostic terms. With Big O, you express complexity in terms of how quickly your algorithms runtime grows relative to the size of the input, especially as the input grows arbitrarily large.

Assuming that n is the size of the input to an algorithm, the Big O notation represents the relationship between n and the number of steps the algorithm takes to find a solution. Big O uses a capital letter O followed by this relationship inside parentheses. For example, O(n) represents algorithms that execute a number of steps proportional to the size of their input.

#### Bubble Sort

Bubble Sort is one of the most straightforward sorting algorithms. Its name comes from the way the algorithm works: With every new pass, the largest element in the list bubbles up toward its correct position.

Bubble sort consists of making multiple passes through a list, comparing elements one by one, and swapping adjacent items that are out of order.

````python
def bubble_sort(array):
    n = len(array)

    for i in range(n):
        # Create a flag that will allow the function to
        # terminate early if there's nothing left to sort
        already_sorted = True

        # Start looking at each item of the list one by one,
        # comparing it with its adjacent value. With each
        # iteration, the portion of the array that you look at
        # shrinks because the remaining items have already been
        # sorted.
        for j in range(n - i - 1):
            if array[j] > array[j + 1]:
                # If the item you're looking at is greater than its
                # adjacent value, then swap them
                array[j], array[j + 1] = array[j + 1], array[j]

                # Since you had to swap two elements,
                # set the `already_sorted` flag to `False` so the
                # algorithm doesn't finish prematurely
                already_sorted = False

        # If there were no swaps during the last iteration,
        # the array is already sorted, and you can terminate
        if already_sorted:
            break

    return array
````

**Big O Runtime Complexity**:

Implementation of bubble sort consists of two nested for loops in which the algorithm performs n - 1 comparisons, then n - 2 comparisons, and so on until the final comparison is done. This comes at a total of (n - 1) + (n - 2) + (n - 3) +  + 2 + 1 = n(n-1)/2 comparisons, which can also be written as n2 - n.

Doing so simplifies the notation to n2 - n. Since n2 grows much faster than n, this last term can be dropped as well, leaving bubble sort with an average- and worst-case complexity of O(n2).

**Strengths and Weaknesses of Bubble Sort**:

 Advantage of the bubble sort algorithm is its simplicity. It is straightforward to both implement and understand. This is probably the main reason why most computer science courses introduce the topic of sorting using bubble sort.

  Disadvantage of bubble sort is that it is slow, with a runtime complexity of O(n2). Unfortunately, this rules it out as a practical candidate for sorting large arrays.

####  Insertion Sort

Insertion sort algorithm is straightforward to implement and understand. But unlike bubble sort, it builds the sorted list one element at a time by comparing each item with the rest of the list and inserting it into its correct position. This insertion procedure gives the algorithm its name.

````python
def insertion_sort(array):
    # Loop from the second element of the array until
    # the last element
    for i in range(1, len(array)):
        # This is the element we want to position in its
        # correct place
        key_item = array[i]

        # Initialize the variable that will be used to
        # find the correct position of the element referenced
        # by `key_item`
        j = i - 1

        # Run through the list of items (the left
        # portion of the array) and find the correct position
        # of the element referenced by `key_item`. Do this only
        # if `key_item` is smaller than its adjacent values.
        while j >= 0 and array[j] > key_item:
            # Shift the value one position to the left
            # and reposition j to point to the next element
            # (from right to left)
            array[j + 1] = array[j]
            j -= 1

        # When you finish shifting the elements, you can position
        # `key_item` in its correct location
        array[j + 1] = key_item

    return array
````

**Big O Runtime Complexity**:

Insertion sort algorithm has a couple of nested loops that go over the list. The inner loop is pretty efficient because it only goes through the list until it finds the correct position of an element. That said, the algorithm still has an O(n2) runtime complexity on the average case.

The worst case happens when the supplied array is sorted in reverse order. In this case, the inner loop has to execute every comparison to put every element in its correct position. This still gives you an O(n2) runtime complexity.

The best case happens when the supplied array is already sorted. Here, the inner loop is never executed, resulting in an O(n) runtime complexity, just like the best case of bubble sort.

**Strengths and Weaknesses**:

Insertion sort is an O(n2) algorithm, its also much more efficient in practice than other quadratic implementations such as bubble sort.

There are more powerful algorithms, including merge sort and Quicksort, but these implementations are recursive and usually fail to beat insertion sort when working on small lists. Some Quicksort implementations even use insertion sort internally if the list is small enough to provide a faster overall implementation. Timsort also uses insertion sort internally to sort small portions of the input array.

That said, insertion sort is not practical for large arrays, opening the door to algorithms that can scale in more efficient ways.

#### Merge Sort

Merge sort is a very efficient sorting algorithm. Its based on the divide-and-conquer approach, a powerful algorithmic technique used to solve complex problems.

Divide-and-conquer algorithms typically follow the same structure:

1. The original input is broken into several parts, each one representing a subproblem thats similar to the original but simpler.
2. Each subproblem is solved recursively.
3. The solutions to all the subproblems are combined into a single overall solution.

In the case of merge sort, the divide-and-conquer approach divides the set of input values into two equal-sized parts, sorts each half recursively, and finally merges these two sorted parts into a single sorted list.

**Implementing Merge Sort in Python**:

The implementation of the merge sort algorithm needs two different pieces:

A function that recursively splits the input in half
A function that merges both halves, producing a sorted array

````python
def merge(left, right):
    # If the first array is empty, then nothing needs
    # to be merged, and you can return the second array as the result
    if len(left) == 0:
        return right

    # If the second array is empty, then nothing needs
    # to be merged, and you can return the first array as the result
    if len(right) == 0:
        return left

    result = []
    index_left = index_right = 0

    # Now go through both arrays until all the elements
    # make it into the resultant array
    while len(result) < len(left) + len(right):
        # The elements need to be sorted to add them to the
        # resultant array, so you need to decide whether to get
        # the next element from the first or the second array
        if left[index_left] <= right[index_right]:
            result.append(left[index_left])
            index_left += 1
        else:
            result.append(right[index_right])
            index_right += 1

        # If you reach the end of either array, then you can
        # add the remaining elements from the other array to
        # the result and break the loop
        if index_right == len(right):
            result += left[index_left:]
            break

        if index_left == len(left):
            result += right[index_right:]
            break

    return result


def merge_sort(array):
    # If the input array contains fewer than two elements,
    # then return it as the result of the function
    if len(array) < 2:
        return array

    midpoint = len(array) // 2

    # Sort the array by recursively splitting the input
    # into two equal halves, sorting each half and merging them
    # together into the final result
    return merge(
        left=merge_sort(array[:midpoint]),
        right=merge_sort(array[midpoint:]))
````

**Big O Complexity**:

To analyze the complexity of merge sort, you can look at its two steps separately:

- merge() has a linear runtime. It receives two arrays whose combined length is at most n (the length of the original input array), and it combines both arrays by looking at each element at most once. This leads to a runtime complexity of O(n).

- The second step splits the input array recursively and calls merge() for each half. Since the array is halved until a single element remains, the total number of halving operations performed by this function is log2n. Since merge() is called for each half, we get a total runtime of O(n log2n).

Interestingly, O(n log2n) is the best possible worst-case runtime that can be achieved by a sorting algorithm.

**Strengths and Weaknesses**:

Thanks to its runtime complexity of O(n log2n), merge sort is a very efficient algorithm that scales well as the size of the input array grows. Its also straightforward to parallelize because it breaks the input array into chunks that can be distributed and processed in parallel if necessary.

That said, for small lists, the time cost of the recursion allows algorithms such as bubble sort and insertion sort to be faster.

Another drawback of merge sort is that it creates copies of the array when calling itself recursively. It also creates a new list inside merge() to sort and return both input halves. This makes merge sort use much more memory than bubble sort and insertion sort, which are both able to sort the list in place.

Due to this limitation, you may not want to use merge sort to sort large lists in memory-constrained hardware.

#### Quicksort

 Quicksort algorithm applies the divide-and-conquer principle to divide the input array into two lists, the first with small items and the second with large items. The algorithm then sorts both lists recursively until the resultant list is completely sorted.

Dividing the input list is referred to as partitioning the list. Quicksort first selects a pivot element and partitions the list around the pivot, putting every smaller element into a low array and every larger element into a high array.

Putting every element from the low list to the left of the pivot and every element from the high list to the right positions the pivot precisely where it needs to be in the final sorted list. This means that the function can now recursively apply the same procedure to low and then high until the entire list is sorted.

````python
from random import randint

def quicksort(array):
    # If the input array contains fewer than two elements,
    # then return it as the result of the function
    if len(array) < 2:
        return array

    low, same, high = [], [], []

    # Select your `pivot` element randomly
    pivot = array[randint(0, len(array) - 1)]

    for item in array:
        # Elements that are smaller than the `pivot` go to
        # the `low` list. Elements that are larger than
        # `pivot` go to the `high` list. Elements that are
        # equal to `pivot` go to the `same` list.
        if item < pivot:
            low.append(item)
        elif item == pivot:
            same.append(item)
        elif item > pivot:
            high.append(item)

    # The final result combines the sorted `low` list
    # with the `same` list and the sorted `high` list
    return quicksort(low) + same + quicksort(high)
````

**Selecting the pivot Element**:

Why does the implementation above select the pivot element randomly? Wouldnt it be the same to consistently select the first or last element of the input list?

Because of how the Quicksort algorithm works, the number of recursion levels depends on where pivot ends up in each partition. In the best-case scenario, the algorithm consistently picks the median element as the pivot. That would make each generated subproblem exactly half the size of the previous problem, leading to at most log2n levels.

On the other hand, if the algorithm consistently picks either the smallest or largest element of the array as the pivot, then the generated partitions will be as unequal as possible, leading to n-1 recursion levels. That would be the worst-case scenario for Quicksort.

As you can see, Quicksorts efficiency often depends on the pivot selection. If the input array is unsorted, then using the first or last element as the pivot will work the same as a random element. But if the input array is sorted or almost sorted, using the first or last element as the pivot could lead to a worst-case scenario. Selecting the pivot at random makes it more likely Quicksort will select a value closer to the median and finish faster.

Another option for selecting the pivot is to find the median value of the array and force the algorithm to use it as the pivot. This can be done in O(n) time. Although the process is little bit more involved, using the median value as the pivot for Quicksort guarantees you will have the best-case Big O scenario.

**Big O Complexity**:

With Quicksort, the input list is partitioned in linear time, O(n), and this process repeats recursively an average of log2n times. This leads to a final complexity of O(n log2n).

That said, remember the discussion about how the selection of the pivot affects the runtime of the algorithm. The O(n) best-case scenario happens when the selected pivot is close to the median of the array, and an O(n2) scenario happens when the pivot is the smallest or largest value of the array.

Theoretically, if the algorithm focuses first on finding the median value and then uses it as the pivot element, then the worst-case complexity will come down to O(n log2n). The median of an array can be found in linear time, and using it as the pivot guarantees the Quicksort portion of the code will perform in O(n log2n).

By using the median value as the pivot, you end up with a final runtime of O(n) + O(n log2n). You can simplify this down to O(n log2n) because the logarithmic portion grows much faster than the linear portion.

> Note: Although achieving O(n log2n) is possible in Quicksorts worst-case scenario, this approach is seldom used in practice. Lists have to be quite large for the implementation to be faster than a simple randomized selection of the pivot.

> Randomly selecting the pivot makes the worst case very unlikely. That makes random pivot selection good enough for most implementations of the algorithm.

**Strengths and Weaknesses**:

True to its name, Quicksort is very fast. Although its worst-case scenario is theoretically O(n2), in practice, a good implementation of Quicksort beats most other sorting implementations. Also, just like merge sort, Quicksort is straightforward to parallelize.

One of Quicksorts main disadvantages is the lack of a guarantee that it will achieve the average runtime complexity. Although worst-case scenarios are rare, certain applications cant afford to risk poor performance, so they opt for algorithms that stay within O(n log2n) regardless of the input.

Just like merge sort, Quicksort also trades off memory space for speed. This may become a limitation for sorting larger lists.

 Quicksort also pays the price of recursion when the list is sufficiently small, taking longer to complete than both insertion sort and bubble sort.

#### Timsort

The Timsort algorithm is considered a hybrid sorting algorithm because it employs a best-of-both-worlds combination of insertion sort and merge sort.

The main characteristic of Timsort is that it takes advantage of already-sorted elements that exist in most real-world datasets. These are called natural runs. The algorithm then iterates over the list, collecting the elements into runs and merging them into a single sorted list.


````python
def timsort(array):
    min_run = 32
    n = len(array)

    # Start by slicing and sorting small portions of the
    # input array. The size of these slices is defined by
    # your `min_run` size.
    for i in range(0, n, min_run):
        insertion_sort(array, i, min((i + min_run - 1), n - 1))

    # Now you can start merging the sorted slices.
    # Start from `min_run`, doubling the size on
    # each iteration until you surpass the length of
    # the array.
    size = min_run
    while size < n:
        # Determine the arrays that will
        # be merged together
        for start in range(0, n, size * 2):
            # Compute the `midpoint` (where the first array ends
            # and the second starts) and the `endpoint` (where
            # the second array ends)
            midpoint = start + size - 1
            end = min((start + size * 2 - 1), (n-1))

            # Merge the two subarrays.
            # The `left` array should go from `start` to
            # `midpoint + 1`, while the `right` array should
            # go from `midpoint + 1` to `end + 1`.
            merged_array = merge(
                left=array[start:midpoint + 1],
                right=array[midpoint + 1:end + 1])

            # Finally, put the merged array back into
            # your array
            array[start:start + len(merged_array)] = merged_array

        # Each iteration should double the size of your arrays
        size *= 2

    return array
````

 unlike merge sort, Timsort merges subarrays that were previously sorted. Doing so decreases the total number of comparisons required to produce a sorted list. This advantage over merge sort will become apparent when running experiments using different arrays.

Finally, line 2 defines min_run = 32. There are two reasons for using 32 as the value here:

Sorting small arrays using insertion sort is very fast, and min_run has a small value to take advantage of this characteristic. Initializing min_run with a value thats too large will defeat the purpose of using insertion sort and will make the algorithm slower.

Merging two balanced lists is much more efficient than merging lists of disproportionate size. Picking a min_run value thats a power of two ensures better performance when merging all the different runs that the algorithm creates.

Combining both conditions above offers several options for min_run. The implementation in this tutorial uses min_run = 32 as one of the possibilities.

> Note: In practice, Timsort does something a little more complicated to compute min_run. It picks a value between 32 and 64 inclusive, such that the length of the list divided by min_run is exactly a power of 2. If thats not possible, it chooses a value thats close to, but strictly less than, a power of 2.



**Timsorts Big O**

On average, the complexity of Timsort is O(n log2n), just like merge sort and Quicksort. The logarithmic part comes from doubling the size of the run to perform each linear merge operation.

However, Timsort performs exceptionally well on already-sorted or close-to-sorted lists, leading to a best-case scenario of O(n). In this case, Timsort clearly beats merge sort and matches the best-case scenario for Quicksort. But the worst case for Timsort is also O(n log2n), which surpasses Quicksorts O(n2).

**Strengths and Weaknesses**

The main disadvantage of Timsort is its complexity. Despite implementing a very simplified version of the original algorithm, it still requires much more code because it relies on both insertion_sort() and merge().

One of Timsorts advantages is its ability to predictably perform in O(n log2n) regardless of the structure of the input array. Contrast that with Quicksort, which can degrade down to O(n2). Timsort is also very fast for small arrays because the algorithm turns into a single insertion sort.

For real-world usage, in which its common to sort arrays that already have some preexisting order, Timsort is a great option. Its adaptability makes it an excellent choice for sorting arrays of any length.

### Distributed Data Processing Frameworks

**Hadoop:**

- Hadoop is an open-source framework for distributed storage and processing of large datasets. It consists of two main components:
    - HDFS (Hadoop Distributed File System): A storage layer that stores data across multiple nodes.
    - MapReduce: A programming model for processing data in parallel across clusters of computers.
- Its typically used for batch processing large volumes of data but can be slower than newer technologies due to its reliance on disk-based processing.


**Spark:**

- Apache Spark is a distributed computing framework that provides faster in-memory processing compared to Hadoop. It supports both batch and real-time processing.
- Spark can run on top of Hadoop and uses Resilient Distributed Datasets (RDDs), which are collections of data that can be processed in parallel.
- It also includes libraries for SQL (Spark SQL), machine learning (MLlib), and graph processing (GraphX), making it versatile for a wide range of data processing tasks.

**Flink:**

- Apache Flink is a stream-processing framework that also supports batch processing. It excels in real-time, low-latency processing and is designed for handling unbounded streams of data.
- Unlike Spark, which was initially focused on batch processing, Flink has strong real-time processing capabilities, making it ideal for complex event processing (CEP) and stream analytics.

### Designing Scalable Data Pipelines

Data pipelines are designed to process, store, and analyze data. They can be batch or real-time based, depending on the applications requirements.

**Batch Processing:**
- In batch processing, data is collected over time and processed in bulk. This approach is suitable for jobs that do not require immediate results.
- Examples include nightly ETL jobs, large-scale data aggregation, and reporting.
- Tools: Hadoop, Spark (with batch mode), Apache Flink (can also handle batch).

**Real-time Processing:**

- Real-time data pipelines process data immediately as it is generated or received. This is crucial for applications that require up-to-date insights.
- Examples include fraud detection systems, live user analytics, and IoT systems.
- Tools: Apache Kafka, Apache Flink, Spark Streaming, AWS Kinesis.


**Scalability Considerations:**

- Both types of pipelines need to be scalable to handle increasing data volumes. The key is to use distributed computing frameworks like Hadoop or Spark, along with scalable data storage (like HDFS or cloud storage solutions).
- For real-time systems, low-latency and the ability to process high-velocity streams are critical.

### Event-Driven Architectures

Event-driven architectures (EDA) are designed around events (such as user actions, sensor readings, etc.) that trigger certain processing actions. These architectures are crucial for real-time systems where events must be processed as they occur.

**Kafka:**

Apache Kafka is a distributed event streaming platform used for building real-time data pipelines. It allows for the production, consumption, and storage of streams of events. Kafka is designed for fault tolerance and high throughput and is commonly used in stream processing and messaging systems.

**Event Hubs:**

Azure Event Hubs is a cloud-based data streaming platform provided by Microsoft. It allows you to ingest massive amounts of data from various sources and stream it to real-time analytics and processing systems like Azure Stream Analytics, Apache Kafka, or custom consumer applications.

**Kinesis:**

Amazon Kinesis is a platform on AWS for building real-time data streaming applications. It can process large streams of data for real-time analytics and integrates with various AWS services like Lambda, Redshift, and S3.

These platforms support publish-subscribe and message queuing models, enabling asynchronous communication and decoupling of components in a distributed system.

### CAP Theorem, Consistency Models, and Trade-offs

#### CAP Theorem:

The CAP Theorem states that a distributed system can only guarantee two out of three properties at any given time:

- Consistency: Every read operation returns the most recent write.
- Availability: Every request (read or write) will receive a response (not necessarily the most recent data).
- Partition Tolerance: The system can continue to function even if some - network partitions occur, i.e., some nodes cannot communicate with others.

In practice, this means that systems have to make trade-offs between these properties. For example:

- CP (Consistency + Partition tolerance): Systems like HBase and Zookeeper fall under this category.
- AP (Availability + Partition tolerance): Systems like Cassandra and DynamoDB are available even when consistency might not be guaranteed.
- CA (Consistency + Availability): While theoretically possible, CA systems are impractical in distributed systems with network partitions.

#### Consistency Models:

Different systems may adopt different consistency models based on how they prioritize data accuracy versus availability:

- Strong Consistency: Guarantees that all clients always see the same data at the same time.
- Eventual Consistency: Guarantees that all replicas will eventually converge to the same value, but not necessarily immediately.
- Causal Consistency: Ensures that events that are causally related are seen by all nodes in the same order.

### Designing for High Availability and Fault Tolerance

#### High Availability (HA)

HA ensures that a system is continuously operational with minimal downtime. To achieve this:

- Redundancy: Replicate data and services across multiple nodes or regions.
- Load Balancing: Distribute incoming requests across multiple instances to avoid bottlenecks.
- Failover Mechanisms: Use mechanisms like automatic failover or leader election to handle node failures.

#### Fault Tolerance

Fault tolerance ensures that a system continues functioning despite failures in some of its components. Techniques include:

- Data Replication: Ensure that data is replicated across multiple nodes or regions (e.g., HDFS, Cassandra).
- Checkpoints: In streaming systems (like Kafka or Flink), checkpoints are used to store the state periodically to recover from failures.
- Graceful Degradation: Allow the system to degrade functionality gracefully when some components fail, instead of crashing entirely.
